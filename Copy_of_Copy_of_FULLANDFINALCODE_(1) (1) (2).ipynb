```
#########################
# run this
#########################
# mtcnn method
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from mtcnn.mtcnn import MTCNN
import cv2

# draw an image with detected objects
def faceDetectionMtcnn(filename):
  
  pixels = pyplot.imread(filename)
  
  detector = MTCNN()
  
  faces = detector.detect_faces(pixels)
  
  face_locations = []
  
  for i in range(len(faces)):
    temp = faces[i]['box']
    t1=faces[i]['box'][0]
    t2=faces[i]['box'][1]
    t3=faces[i]['box'][2]
    t4=faces[i]['box'][3]
    faces[i]['box'][0]=int(t2)
    faces[i]['box'][1]=int(t1+t3)
    faces[i]['box'][2]=int(t2+t4)
    faces[i]['box'][3]=int(t1)
    face_locations.append(tuple(faces[i]['box']))
  return face_locations
  # display faces on the original image

```

```
# method for creating encodings of known faces

import face_recognition
from PIL import Image, ImageDraw
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import glob
from pathlib import Path
ourcount = 0
# from skvideo.io import VideoWriter

def encoding_faces(front_img_filename,img_files_names, person_name):
  print(img_files_names)
  front_img = face_recognition.load_image_file('/content/drive/MyDrive/MIlap/Missing_Person/Front_Images/%s' %front_img_filename)   
  filename1 = ('/content/drive/MyDrive/MIlap/Missing_Person/Front_Images/%s' %front_img_filename)
  print(filename1)
  face_locations= faceDetectionMtcnn(filename1)
  face_encoding = face_recognition.face_encodings(front_img, face_locations)[0]


  side1_img = face_recognition.load_image_file('/content/drive/MyDrive/MIlap/Missing_Person/Side_Images/%s' %img_files_names[0])   
  filename2 = ('/content/drive/MyDrive/MIlap/Missing_Person/Side_Images/%s' %img_files_names[0])
  face_locations= faceDetectionMtcnn(filename2)
  side1_face_encoding = face_recognition.face_encodings(side1_img, face_locations)[0]

  side2_img = face_recognition.load_image_file('/content/drive/MyDrive/MIlap/Missing_Person/Side_Images/%s' %img_files_names[1])    
  filename3 = ('/content/drive/MyDrive/MIlap/Missing_Person/Side_Images/%s' %img_files_names[1])
  face_locations= faceDetectionMtcnn(filename3) 
  side2_face_encoding = face_recognition.face_encodings(side2_img, face_locations)[0]
  


  #  Create arrays of encodings and names
  known_face_encodings = [
    face_encoding,
    side1_face_encoding,
    side2_face_encoding
  ]
  
  known_face_names = [
    person_name,
    person_name,
    person_name    
  ]
  print(known_face_names)
  
  # Load test image to find faces in
  for n in range(3):
    print('checking video %d' %(n+1))
    for filename in glob.glob('/content/drive/MyDrive/MIlap/Frame_Folder%d/*.jpg' %(n+1)):
      matches=[]
      test_image = face_recognition.load_image_file(filename)
      test_image_filename = filename
      output= test_image.copy()

      # Find faces in test image
      face_locations= faceDetectionMtcnn(test_image_filename)
      face_encodings = face_recognition.face_encodings(test_image, face_locations)
      

      # create a list new list of only box from faces wala dictionary


      # Convert to PIL format
      pil_image = Image.fromarray(test_image)
      
      # Create a ImageDraw instance
      draw = ImageDraw.Draw(pil_image)

      # Loop through faces in test image
      for(top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
        
        print( "List of matches: ",matches)
        name = "Unknown Person"

        face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
        best_match_index = np.argmin(face_distances)

        print("index of best match: ", best_match_index)
        if matches[best_match_index]:
          name = known_face_names[best_match_index]


        # Draw box
        draw.rectangle(((left, top), (right, bottom)), outline=(255,255,0))
        # Draw label
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left,bottom - text_height - 10), (right, bottom)), fill=(255,255,0), outline=(255,255,0))
        draw.text((left + 6, bottom - text_height - 5), name, fill=(0,0,0))


      del draw
      # Display image
      pil_image.show()

      # Save image
      
      for i in matches:
        if i:
          global ourcount;
          ourcount = format(ourcount, '05')
          pil_image.save('/content/drive/MyDrive/MIlap/Identify_Folder%d/identifys%s.jpg' % ((n+1),ourcount))
          ourcount = int(ourcount) + 1
          print("theres true")   
  cv2.destroyAllWindows()  
```
